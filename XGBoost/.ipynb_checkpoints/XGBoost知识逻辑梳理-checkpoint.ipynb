# XGBoost模型构造
## 如何集成
* 基础模型：决策树模型——树模型是一个效果非常好的基础模型，用树模型来集成效果一般都会好很多。
* 集成方法：一个一个加【并非简单并行构造多棵树】

* 每一轮加入一个什么样的基础模型？

* 树模型的结构
    1. 叶子个数限制——让树不要太庞大
    2. 权重正则化——限制基础的树模型
* 目标函数
    1. 一方面要保证预测值和真实值之间的差异最小
    2. 另一方面要让树模型更加精简

$\sum{i=1}{n}$

## XGBoost模型
* 目标：$$\mathcal{L}^{(t)}=\sum_{i = 1}^{n}l\left(y_i, \hat{y_i}^{(t-1)}+f_t(x_i)\right)+\Omega(f_t)+constant$$
* 用泰勒展开来近似我们原来的目标
    * 泰勒展开：$$f(x+{\Delta}x){\approx}f(x)+f'(x){\Delta}x+\frac{1}{2}f''(x){\Delta}x^2$$
    * 定义：$$g_i = \partial _{\hat{y}^{(t-1)}}l(y_i,\hat{y}^{(t-1)}),h_i = \partial_{\hat{y}^{(t-1)}}^2l(y_i,\hat{y}^{(t-1)})$$
* 简化后的目标：$$\sum_{i=1}^{n}[g_if_t(x_i)+\frac{1}{2}h_if_t^2(x_i)]+\Omega (f_t)$$
